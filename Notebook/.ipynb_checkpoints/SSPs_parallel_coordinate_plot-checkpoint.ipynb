```python
import numpy as np
import pandas as pd
from pandas.plotting import parallel_coordinates
import matplotlib.pyplot as plt
import seaborn
```

```python
import random


columns = ['Population', 'Education', 'Urbanisation', 'GDP', 'Energy demand', 
                           'Coal production', 'Gas production', 'Oil production', 'Wind production', 
                           'Solar production', 'Biomass production', 'Livestock demand', 'Meat consumption', 
                           'Land productivity', 'Deforestation', 'Emissions']



mylist = [0, 0.25, 0.5, 0.75, 1, 0, 0.25, 0.5, 0.75, 1,0, 0.25, 0.5, 0.75, 1,0, 0.25, 0.5, 0.75, 1, ]

a = []
for i in range(500):
    a.append(random.choices(mylist, k=16))
    
df = pd.DataFrame(a, columns=columns)
df['State of the world'] = 'Plausible states of the world'
```

```python
directory = 'N:/Brett-lab/Enayat/Felix/Felix_SDGs/'
df_ssps = pd.read_excel(directory+'ScenarioFramework.xlsx', sheet_name='Sheet3')


for i in range(5):
    df.iloc[-(i+1)] = df_ssps.iloc[-(i+1)]
```

```python
df
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(13, 5)) 

parallel_coordinates(df,'State of the world', color= ['#dfdfde', '#61b087','#7f58af','#e84d8a', '#64c5eb', '#feb326'], 
                     linewidth=4, alpha=1, ax=ax)

labels = ax.get_xticklabels()
ax.set_xticklabels(labels, rotation=45, ha='right')
plt.yticks([0, 0.25, 0.5, 0.75, 1],('Low', 'Relatively low', 'Medium', 'Relatively high', 'High') )

plt.legend(loc='center', bbox_to_anchor=(.5, 1.1),   ncol=6)

plt.savefig('{}/SSPs_parallel_coordinate.png'.format(r'C:/Users/moallemie/EM_analysis/Fig/'), dpi=300,  bbox_inches='tight') 

plt.show()
```

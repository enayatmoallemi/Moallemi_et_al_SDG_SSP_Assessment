```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sc = 2000    # Specify the number of scenarios where the convergence in the SA indices occures and was used for correlation analysis
t = 2100
directory = 'N:/Planet-A/Current-Users/Enayat-Moallemi/Moallemi_et_al_SDG_SSP_Assessment/Model/'
df_out = pd.read_excel(directory+'ScenarioFramework.xlsx', sheet_name='Outcomes')
top_factors = {"Total Population Indicator": 6, "Total Primary Education Graduates Indicator": 10, 
               "Total Secondary Education Graduates Indicator": 9, "Total Tertiary Education Graduates Indicator": 8, 
               "GWP per Capita Indicator": 5, "Energy Demand Indicator": 7, "Solar Energy Production Indicator": 15, 
               "Wind Energy Production Indicator": 14, "Biomass Energy Production Indicator": 18, 
               "Oil Production Indicator": 20, "Gas Production Indicator": 18, "Coal Production Indicator": 18, 
               "Total Croplands Indicator": 13, "Forest Land Indicator": 13, "Pasture Land Indicator": 14, 
               "Urban Industrial Land Indicator": 8, "Nonenergy Crops Production Indicator": 9, 
               "Livestock Production Indicator": 16, "Total CO2 Emissions Indicator": 20, 
               "CO2 Radiative Forcing Indicator": 8}
outcome_vars = list(top_factors.keys())  # Specify the outcome of interest for SA ranking verification
```

```python
# Create a dataframe of all to 20 uncertain parameters across all output indicators (remove duplicates when uncertin parameters are repeated in multiple output indicators).
# Also, copy the list of important uncertainties, across all output indicators, in a spreadsheet to be used for SSP calibration


# Specify the directory of the SA results
sa_dir = 'N:/Planet-A/Current-Users/Enayat-Moallemi/Moallemi_et_al_SDG_SSP_Assessment/Results/Morris_results/'
     
new_mu_df = pd.DataFrame()
impt_mu = []
for i, outcome_var in enumerate(outcome_vars):    
    mu_df = pd.read_csv(sa_dir+"MorrisIndices_{}_sc{}_t{}_test.csv".format(outcome_var, sc, t))
    mu_df.set_index('Unnamed: 0', inplace=True)
    mu_df.index.names = ['Uncertainty']
    mu_df.rename(columns={'mu_star': outcome_var}, inplace=True)
    mu_df.drop(['mu_star_conf', 'mu', 'sigma'], axis=1, inplace=True)
    mu_df.sort_values(by=[outcome_var], ascending=False, inplace=True)
    mu_df2 = mu_df.head(20)
    impt_mu_df = mu_df.head(top_factors[outcome_var])
    new_mu_df = pd.concat([new_mu_df, mu_df2], axis=1)
    impt_mu = impt_mu + list(impt_mu_df.index)


s = set(impt_mu)
# Bring the reference, min, and max values of important uncertainties from ScenarioFramework 
unc_dir = 'N:/Planet-A/Current-Users/Enayat-Moallemi/Moallemi_et_al_SDG_SSP_Assessment/Model/'
df_unc = pd.read_excel(unc_dir+'ScenarioFramework.xlsx', sheet_name='Uncertainties')
#df_unc['Min'] = df_unc['Min'] + df_unc['Reference'] * 0.75
#df_unc['Max'] = df_unc['Max'] + df_unc['Reference'] * 1.25

impt_unc = pd.DataFrame(index=list(s), columns=['Reference', 'Min', 'Max'])

for u, unc in enumerate(list(impt_unc.index)):
    impt_unc.loc[unc, 'Reference'] = df_unc[df_unc['Uncertainty']==unc]['Reference'].values[0]
    impt_unc.loc[unc, 'Max'] = df_unc[df_unc['Uncertainty']==unc]['Max'].values[0]
    impt_unc.loc[unc, 'Min'] = df_unc[df_unc['Uncertainty']==unc]['Min'].values[0]
    
# Reorder the importrant uncertainties according to the original uncertainty order in ScenarioFramework and remove missing values
impt_unc = impt_unc.reindex(df_unc['Uncertainty'])
impt_unc = impt_unc[impt_unc['Reference'].notna()]
```

```python
# Create a dataframe of all to 20 uncertain parameters across all output indicators (remove duplicates when uncertin parameters are repeated in multiple output indicators).

sc= 3000

# Specify the directory of the SA results
sa_dir = 'C:/Users/moallemie/GitHub/SSPs_SDGs_Assessment/Results/Morris_results/'

new_mu_df = pd.DataFrame()

for t, time in enumerate([2030, 2050]): 
    
    for i, outcome_var in enumerate(outcome_vars):    
        mu_df = pd.read_csv(sa_dir+"MorrisIndices_{}_sc{}_t{}_test.csv".format(outcome_var, sc, time))
        mu_df.set_index('Unnamed: 0', inplace=True)
        mu_df.index.names = ['Uncertainty']
        mu_df.rename(columns={'mu_star': outcome_var+' '+str(time)}, inplace=True)
        mu_df.drop(['mu_star_conf', 'mu', 'sigma'], axis=1, inplace=True)
        mu_df.sort_values(by=[outcome_var+' '+str(time)], ascending=False, inplace=True)
        mu_df2 = mu_df.head(20) # for time step 2030 and 2050, we ploted and annotated all top 20 factors
        new_mu_df = pd.concat([new_mu_df, mu_df2], axis=1)

new_mu_df = new_mu_df.fillna(0)
normalized_df=(new_mu_df-new_mu_df.min())/(new_mu_df.max()-new_mu_df.min())

uncertainties = normalized_df.index.values.tolist()
labels = pd.DataFrame(index=list(normalized_df.columns.values) , columns=uncertainties)
labels = labels.replace(np.nan, '', regex=True)

# Here annotate all top 20 factors rather than those identified through correlation analysis. We didn't run the correlation analsysi fro 2030 and 2050 time.
for i, outcome_var in enumerate(list(normalized_df.columns.values)):
    topfactor = list(normalized_df.nlargest(20, outcome_var).index)
    for j, factor in enumerate(topfactor):
        labels.loc[outcome_var, factor] = str(j+1)
normalized_df_T = normalized_df.transpose()

labels_year = {}
normalized_df_T_year = {}

labels_year[2030] = labels[0:20]
labels_year[2050] = labels[20:40]

normalized_df_T_year[2030] = normalized_df_T[0:20]
normalized_df_T_year[2050] = normalized_df_T[20:40]
```
